```python
def convert_location_list(loclist):
    ret = []
    for loc in loclist:
        ret.append({
            k.replace("E7", ""): v / 10**7
            for k, v in loc.items()
        })
#     return ret
    return geometry.LineString([geometry.Point(i['lng'], i['lat']) for i in ret])
        
def convert_point(point):
    return geometry.Point(point[1] // 10**7, point[0] // 10**7)
```
